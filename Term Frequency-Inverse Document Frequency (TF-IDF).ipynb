# TF-IDF

how important a word is to a document in a collection or corpus.


term frequency can be said as what is the probability of finding a word in a document



term frequency = No of repetition of words in a sentence / no of words in a sentence


The inverse document frequency is a measure of how much information the word provides



IDF= log(no of sentences / no of sentences contatining words)



TF-IDF = terfm frequency * IDF





```python
import nltk
```

```python
paragraph =  """Pandit Nehru was born on 14th November 1889 in Allahabad (now Praygraj). Furthermore, his father was Motilal Nehru who was a lawyer by occupation. And he was very rich because of which he got the best education.

Moreover, he was sent abroad for studies at an early age. In England, he got his education from two universities namely Cambridge and Harrow. In 1910, Nehruji completed his degree.

He was an average guy in his studies and does not have much interest in studying law. Instead, he had an interest in politics. However, later on, he becomes a lawyer and practiced law in Allahabad High Court. He married to Smt. Kamla Devi at the age of 24. Soon after, they gave birth to a daughter whom they named Indira."""
```

```python
# Cleaning the texts
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
wordnet=WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
sentences
```

```python
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    review = review.split()
    review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
stopwords.words('english')
```

```python
# Creating the TF-IDF model
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
X = cv.fit_transform(corpus).toarray()
```
